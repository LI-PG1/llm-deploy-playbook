## 一、项目简介

### 1.1 官方项目简介

gte-Qwen2-1.5B-instruct 是阿里巴巴NLP团队开源的 **1.5B 参数多语言文本嵌入模型**，基于 Qwen2-1.5B 底座构建，是 gte（General Text Embedding）系列最新模型。

核心特性：
- 双向注意力机制，增强上下文理解
- 指令调优（Instruction Tuning），query侧支持任务描述
- 32K token 最大输入长度
- 1536 维嵌入向量输出
- 多语言支持（中文、英文、法文、波兰文等）
- APACHE 2.0 许可证
- MTEB(英文) 67.16 / C-MTEB(中文) 67.65

### 1.2 本项目简介

本项目在趋动云平台复现了 gte-Qwen2-1.5B-instruct，提供一键启动的 Gradio Embedding WebUI，支持文本向量化、文本相似度计算、文档检索三大功能。

### 1.3 项目实现类型

推理+WebUI（Gradio Embedding界面）

### 4.2 功能说明

本 WebUI 提供三大功能标签页：

**Tab 1: Text Embedding（文本向量化）**
1. 在「Input Text」输入要编码的文本
2. 可选填写「Task Description」添加任务描述
3. 点击「Generate Embedding」生成 1536 维 L2 归一化向量

**Tab 2: Text Similarity（文本相似度）**
1. 在「Text 1」和「Text 2」分别输入两段文本
2. 可选填写「Task Description」用于 query 侧指令调优
3. 点击「Compute Similarity」查看余弦相似度分数和等级

**Tab 3: Document Search（文档检索）**
1. 在「Search Query」输入查询文本，可选填「Task Description」
2. 在「Documents」文本框粘贴多篇文档（每行一篇）
3. 调整「Top-K Results」选择返回结果数
4. 点击「Search」查看按相似度排序的检索结果

### 4.3 Task Description（任务描述）说明

gte-Qwen2-1.5B-instruct 支持 query 端指令调优。添加合适的任务描述可以提升检索精度：

| 场景 | 推荐 Task Description |
|------|------|
| 搜索/检索 | Given a web search query, retrieve relevant passages that answer the query |
| 语义相似度 | Given a sentence, find semantically similar sentences |
| 分类 | Given a text, classify it into categories |
| 聚类 | Given a text, find similar texts for clustering |
| 问答 | Given a question, retrieve relevant passages that answer the question |

## 六、常见问题

### Q: 模型需要多少显存？
A: BF16 精度下约 3GB，4GB 显存 GPU 即可运行。批量处理多文档时建议 8GB。

### Q: 支持哪些语言？
A: 支持多语言，包括中文、英文、法文、波兰文等，MTEB 多语言评测表现优异。

### Q: 如何提高检索精度？
A: 在「Task Description」中填写合适的任务描述，模型会根据任务描述调整 query 编码方式。

### Q: 文档检索支持多少篇文档？
A: 无硬性限制，但建议一次检索不超过 50 篇以获得最佳响应速度。

## 七、技术架构

```
用户输入文本 → Gradio Embedding WebUI (:7860)
    ↓
AutoTokenizer + AutoModel (trust_remote_code=True)
    ↓
Last Token Pooling + L2 Normalization
    ↓
1536维嵌入向量 / 相似度分数 / 检索结果
```

- 推理引擎：PyTorch 2.5.1 + Transformers 4.46.3
- 模型参数：1.5B（BF16，约 3GB）
- 嵌入维度：1536
- 最大输入：32K tokens
- 池化方式：Last Token Pooling
- 归一化：L2 Normalization
- Web 框架：Gradio 4.7.1
- Python 版本：3.10
- 基础镜像：CUDA 12.1 / PyTorch 2.5.1 / Ubuntu 22.04
