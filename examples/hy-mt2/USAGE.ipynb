# Hy-MT2-7B 多语言翻译模型
## 腾讯混元 · 33语言互译 · Gradio WebUI

---


## 一、项目简介

### 1.1 官方模型

Hy-MT2 是腾讯混元于 2026年5月21日 开源的"快思考"多语言翻译模型家族，涵盖 1.8B、7B 和 30B-A3B（MoE）三种体量，支持 **33种语言** 互译（含粤语、藏语、维吾尔语等方言）。

7B 版本在 FLORES-200 测试中达到 Gemini 2.5 Pro 约 96.9% 的翻译表现，WMT26 翻译竞赛中击败 DeepSeek-V4-Pro 和 Kimi K2.6。

### 1.2 本项目

- **实现类型**：推理 + WebUI（Gradio 翻译界面）
- **模型**：Tencent/Hy-MT2-7B（bfloat16，~14GB）
- **推理引擎**：Transformers 5.6.0 + PyTorch 2.5.1 + CUDA 12.1
- **许可证**：Apache 2.0 开源，可商用

## 五、使用说明

### 5.1 一键启动

```bash
cd /gemini/code
bash start.sh
```

### 5.2 访问地址

点击平台 Web 服务，打开 **端口 7860**

### 5.3 操作步骤

1. 选择源语言（Source Language）
2. 选择目标语言（Target Language）
3. 输入待翻译文本
4. 点击「🌐 翻译」按钮
5. 查看右侧翻译结果

提示：点击 ⇄ 按钮可快速交换源语言和目标语言。


### 5.4 高级参数

展开「⚙️ 高级参数」可调节：

| 参数 | 默认值 | 说明 |
|------|:--:|------|
| Temperature | 0.7 | 越高输出越随机 |
| Top-P | 0.6 | 核采样阈值 |
| Top-K | 20 | 候选词数量 |
| Repetition Penalty | 1.05 | 重复惩罚系数 |
| Max New Tokens | 4096 | 最大生成长度 |
