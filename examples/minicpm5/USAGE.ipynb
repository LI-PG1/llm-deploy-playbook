## 一、项目简介

### 1.1 官方项目简介

MiniCPM5-1B 是 OpenBMB 开源的 **1B 参数端侧大模型**，1B 级别开源 SOTA。采用标准 LlamaForCausalLM 架构，无需自定义代码。

核心特性：
- Think/No-Think 混合推理双模式
- 131K 上下文长度
- 标准 Llama 架构，零定制代码
- BF16 精度，模型仅 2GB
- Apache 2.0 许可证

### 1.2 本项目简介

本项目在趋动云平台复现了 MiniCPM5-1B，提供一键启动的 Gradio Chat WebUI，支持 Think/No-Think 双模式切换。

### 4.2 使用步骤

1. 点击「一键部署」启动项目
2. 模型已预置在镜像中，首次启动自动加载（约 30 秒）
3. 在文本框中输入问题
4. 可勾选「Enable Thinking Mode」启用思考推理模式
5. 调整 Temperature（No Think: 0.7 / Think: 0.9）
6. 按 Enter 或点击 Send 发送

### 4.3 生成参数推荐

| 模式 | Temperature | Top-P | Max Tokens |
|------|:--:|:--:|:--:|
| No Think（通用对话） | 0.7 | 0.95 | 512 |
| Think（数学/代码） | 0.9 | 0.95 | 2048 |
| 确定性（翻译/事实） | 0 | — | 512 |

## 六、常见问题

### Q: 为什么简单数学题也算错？
A: 勾选「Enable Thinking Mode」即可启用 `<think>` 推理链，数学能力大幅提升。

### Q: 如何切换中英文？
A: 模型自动识别输入语言，中英文均可直接提问。

### Q: 显存需要多少？
A: 模型仅 2GB，4GB 显存 GPU 即可流畅运行。

## 七、技术架构

```
用户输入文本 → Gradio Chat WebUI (:7860)
    ↓
AutoModelForCausalLM (LlamaForCausalLM)
    ↓
PyTorch 2.5.1 + CUDA 12.1
    ↓
输出对话文本
```

- 推理引擎：PyTorch + Transformers 5.6.0
- 模型参数：1B（679M 非嵌入参数）
- 模型体积：2GB
- Web 框架：Gradio
- Python 版本：3.10
- 基础镜像：PyTorch 2.5.1 / CUDA 12.1 / Ubuntu 22.04
