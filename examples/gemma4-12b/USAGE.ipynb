## 一、项目简介

### 1.1 官方项目简介

Gemma 4 12B-it 是 Google 推出的多模态大模型，采用 **Gemma4Unified（encoder-free）** 架构，原生支持文本、图像、音频、视频四模态输入，256K 上下文窗口。

核心特性：
- **11.96B 参数**，BF16 精度约 24GB 显存
- **Unified 架构**，无需独立视觉编码器，多模态特征统一处理
- **四模态支持**：Text + Image + Audio + Video
- **原生 256K 上下文长度**
- **Thinking Mode**：支持链式思考，增强复杂推理能力

| 属性 | 值 |
|------|-----|
| 参数量 | 11.96B |
| 架构 | Gemma4Unified |
| 精度 | BF16（约 24GB） |
| 上下文 | 256K |
| 多模态 | Text + Image + Audio + Video |
| 许可证 | Apache-2.0 |

### 1.2 本项目简介

趋动云平台一键部署，Gradio Chat WebUI，支持多模态输入与对话、Thinking Mode 切换、生成参数调节。

| 属性 | 值 |
|------|-----|
| 推理框架 | PyTorch 2.6.0 + Transformers 5.13.1 |
| Web UI | Gradio 6.20.0 |
| Python | 3.10 | CUDA | 12.4 |
| 特色 | 四模态输入、Thinking Mode、参数可调 |

## 四、使用说明

### 4.1 一键启动命令

```bash
# 设置模型路径（已上传至趋动云存储）
export MODEL_PATH=/gemini/pretrain/gemma-4-12B-it-model/

# 一键启动
bash /gemini/code/start.sh
```

### 4.2 使用步骤

1. 开放端口
3. 执行一键启动命令
5. 在输入框中输入文本，或上传图片/音频/视频/文件
6. 按 Enter 或点击「发送」获取回答
7. 可通过左侧开关切换 **Thinking Mode**（链式思考）
8. 「清空对话」重置聊天历史

### 4.3 输入格式说明

| 模态 | 上传方式 | 示例 |
|------|---------|------|
| 文本 | 直接输入 | "写一个if语句" |
| 图像 | 点击「上传图片」 | JPG/PNG |
| 音频 | 点击「上传音频」 | MP3/WAV |
| 视频 | 点击「上传视频」 | MP4 |
| 文档 | 点击「上传文件」 | TXT/PDF 等 |

> 支持单次上传一个文件，发送后自动清空。多模态输入时，在文本框中描述需求并附加上传文件即可。

### 4.4 生成参数推荐

| 场景 | Temperature | Thinking Mode |
|------|:-----------:|:-------------:|
| 通用对话 | 0.7 | 关 |
| 逻辑推理 | 0.5 | 开 |
| 创意写作 | 0.9 | 关 |
| 图像描述 | 0.6 | 关 |
| 精确回答 | 0.2 | 开 |

## 六、模型测评

本模型已完成性能基准评测，涵盖文本、图像、音频、视频、文档共 8 个场景。

### 6.1 评测方案

- 文本 4 档：短（约50 字）、中（约500 字）、长（约2000 字）、超长（约8000 字）
- 多模态 4 项：图像、音频、视频、文档
- 每场景重复 5 次，测量 TTFT、E2E、吞吐量（tok/s）、编码耗时
- 设备：A100 40GB × 1

### 6.2 评测参数

| 参数 | 值 |
|------|:--:|
| Temperature | 0.7 |
| Max New Tokens | 4096（兜底上限） |
| 重复次数 | 5 |
| 数据类型 | BF16 |

### 6.3 结果概览

| 指标 | 值 |
|------|:--:|
| 测试场景数 | 8 场景 × 5 次 |
| 文本 TTFT | ~95ms |
| 图像 TTFT | ~156ms |
| 音频 TTFT | ~217ms（中位数） |
| 视频 TTFT | ~1.5s |
| 平均吞吐 | ~11.5 tok/s |
