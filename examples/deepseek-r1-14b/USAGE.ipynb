## 一、项目简介

### 1.1 官方项目简介

DeepSeek-R1-Distill-Qwen-14B 将 671B MoE 推理大模型的推理能力蒸馏到 Qwen2-14B，生成最终答案前进行 Chain-of-Thought 推理，数学/编程/逻辑表现优异。

核心特性：
- 14B 参数，BF16 精度
- 思维链推理，思考过程可视化
- Qwen2 标准架构，MIT 许可证

| 属性 | 值 |
|------|-----|
| 参数量 | 14B |
| 架构 | Qwen2 |
| 精度 | BF16 (~27.5GB) |
| 上下文 | 131K |
| 许可证 | MIT |

### 1.2 本项目简介

趋动云平台一键部署，Gradio Chat WebUI，支持思维链显示/隐藏、多轮对话。

| 属性 | 值 |
|------|-----|
| 推理框架 | PyTorch 2.5.1 + Transformers 5.9.0 |
| Web UI | Gradio 4.7.1 |
| Python | 3.10 | CUDA | 12.1 |
| 特色 | 思维链切换、多轮对话、System Prompt |

## 四、使用说明

### 4.1 一键启动命令


```python
! bash /gemini/code/start.sh
```

### 4.2 使用步骤

1. 开放 **7860** 端口
2. 执行一键启动命令
3. 等待模型加载（首次约 3 分钟）
4. 输入问题，按 Enter 或点击 Send
5. 勾选/取消 Show Thinking Process 切换思维链显示
6. Clear Conversation 清空对话历史
