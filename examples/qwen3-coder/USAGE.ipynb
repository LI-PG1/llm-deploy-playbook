## 一、项目简介

### 1.1 官方项目简介

Qwen3-Coder-30B-A3B-Instruct 是 Qwen3 系列的代码生成专用模型，采用 MoE（混合专家）架构，总参数量 30.5B，每次推理仅激活 3.3B 参数，兼顾高性能与高效推理。原生支持 256K 上下文窗口，能够处理超长代码文件和复杂项目级任务。

核心特性：
- 30.5B 总参数 / 3.3B 激活参数（MoE 架构）
- 原生 256K 上下文长度
- 专为代码生成与理解优化
- BF16 精度推理，约 61GB 显存需求

| 属性 | 值 |
|------|-----|
| 参数量 | 30.5B（激活 3.3B） |
| 架构 | Qwen3 MoE |
| 精度 | BF16（约 61GB） |
| 上下文 | 256K |
| 许可证 | Apache-2.0 |

### 1.2 本项目简介

趋动云平台一键部署，Gradio Chat WebUI，支持代码生成对话、可调节生成参数。

| 属性 | 值 |
|------|-----|
| 推理框架 | PyTorch 2.6.0 + Transformers 5.13.1 |
| Web UI | Gradio 6.20.0 |
| Python | 3.10 | CUDA | 12.4 |
| 特色 | 256K 超长上下文、代码生成优化、参数可调 |

## 四、使用说明

### 4.1 一键启动命令

```python
! bash /gemini/code/start.sh
```

### 4.2 使用步骤

1. 开放 **7860** 端口
2. 将模型文件上传至 `/gemini/pretrain/Qwen3-Coder-30B-A3B-Instruct-model/`
3. 执行一键启动命令
4. 等待模型加载（首次约 3-5 分钟）
5. 输入代码问题，按 Enter 或点击「发送」
6. 可通过右侧参数面板调整 Temperature、Top-P、Top-K 等参数
7. 「清空对话」重置聊天历史

## 六、模型测评

本模型已完成全面评测，涵盖性能跑分和代码能力评估两大模块。

### 6.1 评测方案

**Part A：性能跑分**
- 4 种代码场景：短代码（约 50 字）、中代码（约 500 字）、长代码（约 2000 字）、超长代码（约 8000 字）
- 每场景重复 3 次，测量 TTFT（首 Token 延迟）、E2E（端到端耗时）、吞吐量（tok/s）
- 输出长度：512 tokens

**Part B：代码能力评估**
- 30 道题，覆盖 6 个核心维度，每维度 5 题
- 维度：基础算法、复杂系统、Agentic 编程、多语言转换、代码审查、测试生成
- 输出长度：8192 tokens
- 评分标准参考附录《代码能力评分计量标准》

### 6.2 评测参数

| 参数 | 值 |
|------|:--:|
| Temperature | 0.7 |
| Top-P | 0.8 |
| Part A 重复次数 | 3 |
| Part B 最大输出 | 8192 tokens |
