# Qwen3.6-35B-A3B

## 一、项目简介

### 1.1 官方项目简介

Qwen3.6-35B-A3B 是阿里巴巴通义千问团队开源的 **MoE（混合专家）架构**大语言对话模型，总参数量 35B，每次推理激活约 3B 参数，Apache 2.0 许可证。

| 属性 | 值 |
|------|-----|
| 总参数量 | 35B |
| 激活参数 | 3B（MoE 架构） |
| 上下文长度 | 262K tokens |
| 模型格式 | FP8 量化 Safetensors（42 分片） |
| 显存需求 | 80GB（P1.large，含 CPU 卸载） |
| 许可证 | Apache 2.0 |

### 1.2 本项目简介

本项目在趋动云平台部署 Qwen3.6-35B-A3B 推理服务，提供 Gradio WebUI 对话界面。支持 Thinking Mode 思维链推理。模型懒加载，首次对话约 48 分钟加载完成后即可流畅对话（~10 tok/s）。

| 属性 | 值 |
|------|-----|
| 推理框架 | PyTorch 2.5.1 + Transformers 5.9.0 |
| Web UI | Gradio 4.7.1 |
| Python | 3.10 |
| CUDA | 12.1（orion-cuda-12.2） |
| 推理速度 | ~10 tok/s（P1.large） |

## 三、算力推荐

| 规格 | 显存 | 推理速度 | 说明 |
|------|:--:|:--:|------|
| P1.large | 80GB | ~10 tok/s | CPU 卸载部分层，性价比高 |
| P1.2xlarge | 160GB | 更快 | 双卡免卸载，模型全放 GPU |

推荐首选 P1.2xlarge（免 CPU 卸载，推理速度最优），P1.large 也已验证可用。


## 四、使用说明

### 推理服务（公网直达）

打开公网地址  输入消息  等待模型加载（首次约 48 分钟） 对话。

- 勾选「Enable Thinking Mode」启用思维链推理（默认开启）
- 可自定义 System Prompt（角色设定）
- 可调 Temperature、Top-P、Max New Tokens 等参数

### 开发环境（一键启动）

`ash
cd /gemini/code && bash start.sh
`
